### Missing Value Treatment Strategy

#### 1. Air Quality Dataset
*   **Xylene:** I am choosing to **Drop the Column**. With over 60% of the data missing, any imputation would essentially be a fabrication. The column is too 'holey' to provide reliable signal for a model.
*   **PM10, NH3, Benzene, Toluene:** I will use **Forward Fill (ffill)**. Air quality data is a time series; the pollution level an hour ago is usually a very strong indicator of the level now. This preserves the local trend better than a global mean.
*   **AQI & AQI_Bucket:** I will **Drop Rows**. Since these are likely our target variables (what we want to predict), imputing them would mean training our model on its own 'guesses', which leads to bias.

#### 2. Crop Production Dataset
*   **Production:** I will **Drop Rows**. Production is the target variable here. Because agricultural yield varies wildly based on unknown factors like local rainfall or soil quality, filling it with a mean or median would introduce too much noise into our training set.

```python
print("Null counts before cleaning (Air Quality):")
print(airquality[['Xylene', 'PM10', 'AQI']].isnull().sum())

airquality_cleaned = airquality.drop(columns=['Xylene'])

airquality_cleaned[['PM10', 'NH3', 'Benzene', 'Toluene']] = airquality_cleaned[['PM10', 'NH3', 'Benzene', 'Toluene']].ffill()

airquality_cleaned = airquality_cleaned.dropna(subset=['AQI', 'AQI_Bucket'])

print("\nNull counts after cleaning (Air Quality):")
print(airquality_cleaned[['PM10', 'AQI']].isnull().sum())
print(f"Xylene in columns: {'Xylene' in airquality_cleaned.columns}")
```

In the code above, I first checked the counts to confirm the problem. Then, I used `drop` to remove the Xylene column entirely because it was mostly empty space. I used the `ffill()` method for the chemical pollutants because environmental data usually changes gradually. Finally, I used `dropna` specifically on the AQI columns to ensure our 'answer key' for the model is 100% real data.

```python
print("Null counts before cleaning (Crop):")
print(cropproduction.isnull().sum())

cropproduction_cleaned = cropproduction.dropna(subset=['Production'])

print("\nNull counts after cleaning (Crop):")
print(cropproduction_cleaned.isnull().sum())
```

For the crop data, the solution was simpler. Since only the 'Production' column had missing values, and that is what we are trying to predict, I removed the rows where that data was missing. This ensures the model only learns from confirmed historical yields.

```python
pollutants_to_fill = ['PM2.5', 'NO', 'NO2', 'NOx', 'SO2', 'O3']
airquality_cleaned[pollutants_to_fill] = airquality_cleaned[pollutants_to_fill].ffill()

airquality_cleaned = airquality_cleaned.dropna(subset=pollutants_to_fill)

print("Final Null Check - Air Quality Cleaned:")
display(airquality_cleaned.isnull().sum())

print("\nFinal Null Check - Crop Production Cleaned:")
display(cropproduction_cleaned.isnull().sum())
```

I applied the forward fill method to the remaining chemical columns like PM2.5, NO, and SO2. This is a common practice for time-series data because environmental factors usually don't jump drastically from one moment to the next.

I also added a final `dropna` on those specific columns just in case there were missing values at the very beginning of the dataset where a 'forward' value didn't exist to pull from. As you can see from the final counts, we now have a solid, clean foundation for both datasets without any lingering empty cells.

I created a mapping dictionary to assign a State to each City in the air quality data. This is crucial because the crop data uses states as its primary geographic marker. By looking at the unique lists, we can see if names like 'Odisha' vs 'Orissa' or 'Tamil Nadu' with extra spaces exist.

```python
def clean_state_names(df, col):
    df[col] = df[col].str.strip().str.title()
    replacement_map = {
        'Orissa': 'Odisha',
        'Telagana': 'Telangana',
        'Andaman And Nicobar Islands': 'Andaman and Nicobar Islands'
    }
    df[col] = df[col].replace(replacement_map)
    return df

airquality_cleaned = clean_state_names(airquality_cleaned, 'State')
cropproduction_cleaned = clean_state_names(cropproduction_cleaned, 'State_Name')

print("Air Quality duplicates before:", airquality_cleaned.duplicated().sum())
print("Crop Production duplicates before:", cropproduction_cleaned.duplicated().sum())

airquality_final = airquality_cleaned.drop_duplicates()
cropproduction_final = cropproduction_cleaned.drop_duplicates()

print("\n--- Record Counts After Cleaning ---")
print(f"Air Quality: {len(airquality_final)} rows")
print(f"Crop Production: {len(cropproduction_final)} rows")
```

I wrote a helper function to standardize the text by stripping extra spaces and converting names to 'Title Case'. I also manually fixed common naming inconsistencies like 'Orissa' vs 'Odisha'. Finally, I used `drop_duplicates` to ensure that identical rows don't skew our future analysis, which is a common issue when data is collected from multiple sensors or reports.

<b>Task 5
Something is making the average AQI look worse than it is

A data quality report flags that a few AQI readings in the dataset are implausibly high — values that no monitoring station should realistically record. If left in, these values will distort every statistic and model built on this data. The team lead asks you to "handle the extreme values properly" but does not tell you how.
Identify whether extreme values exist, quantify them, decide on an appropriate treatment method, apply it, and demonstrate that the data is now cleaner. Justify every decision you make.

The method you chose to detect extremes and why
The count of values affected and the treatment applied
A visual comparison of the data before and after treatment</b>


```python
Q1 = airquality_final['AQI'].quantile(0.25)
Q3 = airquality_final['AQI'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers_count = len(airquality_final[airquality_final['AQI'] > upper_limit])

airquality_no_outliers = airquality_final.copy()
airquality_no_outliers['AQI'] = airquality_no_outliers['AQI'].clip(lower=lower_limit, upper=upper_limit)

print(f"Lower Limit: {lower_limit}")
print(f"Upper Limit: {upper_limit}")
print(f"Number of extreme values treated: {outliers_count}")
```

### Why use the IQR Method?

I chose the **IQR (Interquartile Range) method** because it doesn't rely on the data being perfectly bell-shaped (normal distribution). By calculating the 25th and 75th percentiles, we define a 'normal' range for Indian cities. Any reading above the upper limit is likely a sensor error or a once-in-a-decade event that would confuse a general machine learning model.

### Treatment: Capping vs. Deletion

I decided to **Cap (Winsorize)** the values rather than delete them. If we delete the rows, we lose the information for the other pollutants (like NO2 or CO) recorded on that same day. By capping the AQI at the upper limit, we acknowledge the pollution was very high but prevent 'impossible' numbers from ruining our math.
