```python
import pandas as pd
import os

# Define the paths from the KaggleHub downloads
air_quality_data_dir = "/kaggle/input/air-quality-data-in-india"
crop_production_data_dir = "/root/.cache/kagglehub/datasets/abhinand05/crop-production-in-india/versions/1"
```

```python
air_quality_files = [f for f in os.listdir(air_quality_data_dir) if f.endswith('.csv')]

if air_quality_files:
    air_quality_csv_path = os.path.join(air_quality_data_dir, air_quality_files[0])
    airquality = pd.read_csv(air_quality_csv_path)
    display(airquality.head())
else:
    print(f"No CSV files found in {air_quality_data_dir}")
```

```python
crop_production_files = [f for f in os.listdir(crop_production_data_dir) if f.endswith('.csv')]

if crop_production_files:
    crop_production_csv_path = os.path.join(crop_production_data_dir, crop_production_files[0])
    cropproduce = pd.read_csv(crop_production_csv_path)
    display(cropproduce.head())

else:
    print(f"No CSV files found in {crop_production_data_dir}")
```

### Data Profile for `airquality` Dataset

```python
print("--- airquality Dataset Profile ---")
print(f"Shape: {airquality.shape[0]} rows, {airquality.shape[1]} columns")
print("\nColumn Information:")
airquality.info()
print("\nMissing Values:")
missing_airquality = airquality.isnull().sum()
missing_airquality_pct = (airquality.isnull().sum() / len(airquality)) * 100
missing_airquality_df = pd.DataFrame({'Missing Count': missing_airquality, 'Missing Percentage': missing_airquality_pct})
display(missing_airquality_df[missing_airquality_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False))
print("\nDescriptive Statistics:")
display(airquality.describe())
print("\nUnique Values for Categorical/Object Columns:")
for col in airquality.select_dtypes(include='object').columns:
    print(f"- {col}: {airquality[col].nunique()} unique values")
    if airquality[col].nunique() < 20:
        print(airquality[col].value_counts())
    else:
        print(airquality[col].value_counts().head(10))
duplicates_airquality = airquality.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates_airquality}")
```

### Data Profile for `cropproduce` Dataset

```python
print("--- cropproduce Dataset Profile ---")
print(f"Shape: {cropproduce.shape[0]} rows, {cropproduce.shape[1]} columns")
print("\nColumn Information:")
cropproduce.info()
print("\nMissing Values:")
missing_cropproduce = cropproduce.isnull().sum()
missing_cropproduce_pct = (cropproduce.isnull().sum() / len(cropproduce)) * 100
missing_cropproduce_df = pd.DataFrame({'Missing Count': missing_cropproduce, 'Missing Percentage': missing_cropproduce_pct})
display(missing_cropproduce_df[missing_cropproduce_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False))
print("\nDescriptive Statistics:")
display(cropproduce.describe())
print("\nUnique Values for Categorical/Object Columns:")
for col in cropproduce.select_dtypes(include='object').columns:
    print(f"- {col}: {cropproduce[col].nunique()} unique values")
    if cropproduce[col].nunique() < 20:
        print(cropproduce[col].value_counts())
    else:
        print(cropproduce[col].value_counts().head(10))
duplicates_cropproduce = cropproduce.duplicated().sum()
print(f"\nNumber of Duplicate Rows: {duplicates_cropproduce}")
```

```python
print("airquality: Null Count")
display(airquality.isnull().sum()[airquality.isnull().sum() > 0])


airquality_cleaned = airquality.drop(columns=['Status'])

airquality = airquality_cleaned
```

```python
print("cropproduce-Null Count")
display(cropproduce.isnull().sum()[cropproduce.isnull().sum() > 0])

cropproduce_cleaned = cropproduce.dropna(subset=['Production'])

cropproduce = cropproduce_cleaned
```

```python
print("Airquality Null Counts After Treatment:")
display(airquality.isnull().sum())

print("\nCropproduce Null Counts After Treatment:")
display(cropproduce.isnull().sum())
```

### Rationale for Missing Value Treatment

**`airquality` Dataset:**
*   **`Status` column:** Dropped due to a very high percentage of missing values (42.17%). Imputing such a large portion of a categorical variable would likely introduce significant bias or noise.

**`cropproduce` Dataset:**
*   **`Production` column:** Rows with missing `Production` values were dropped. As 'Production' is a key numerical feature (potentially a target variable), dropping rows with missing values (1.5% of the data) preserves data integrity without introducing artificial values through imputation. This approach is acceptable given the low percentage of missing data.
